```python
import pandas as pd
```

```python
dataframe.drop(['MouseID'],axis=1,inplace=True)
dataframe.drop(['Genotype'],axis=1,inplace=True)
dataframe.drop(['Treatment'],axis=1,inplace=True)
dataframe.drop(['Behavior'],axis=1,inplace=True)
dataframe.drop(['class'],axis=1,inplace=True)
```

```python
dataframe.info()
```

```python
dataframe.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

Imputing missing values

```python
mean_imputer = SimpleImputer(strategy='mean')

for column in dataframe.columns:
  if dataframe[column].isnull().sum() > 0:
    dataframe[column] = mean_imputer.fit_transform(dataframe[column].values.reshape(-1,1))
```

```python
dataframe.isnull().sum()
```

```python
from sklearn.decomposition import PCA
```

```python
feature_columns = dataframe.columns
```

```python
feature_columns
```

```python
from matplotlib import pyplot as plt
import numpy as np
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

from mpl_toolkits.mplot3d import Axes3D

import seaborn as sns
```

```python
pca = PCA(n_components=3)
pca_result = pca.fit_transform(dataframe[feature_columns].values)

#dataframe_with_pca = dataframe

dataframe['pca-one'] = pca_result[:,0]
dataframe['pca-two'] = pca_result[:,1] 
dataframe['pca-three'] = pca_result[:,2]

print('Explained variation per principal component: {}'.format(pca.explained_variance_ratio_))
```

Ð’Ð¸Ð·ÑƒÐµÐ»Ð¸Ð·Ð°Ñ†Ð¸Ñ˜Ð° Ð²Ð¾ 3Ð” ÑÐ¾ PCA

```python
ax = plt.figure(figsize=(16,10)).add_subplot(projection='3d')
ax.scatter(
    xs=dataframe["pca-one"], 
    ys=dataframe["pca-two"], 
    zs=dataframe["pca-three"], 
    cmap='tab10'
)
ax.set_xlabel('pca-one')
ax.set_ylabel('pca-two')
ax.set_zlabel('pca-three')
plt.show()
```

```python
n_components = np.arange(1, 21)
models = [GaussianMixture(n, covariance_type='full', random_state=0).fit(dataframe) for n in n_components]
plt.plot(n_components, [m.bic(dataframe) for m in models], label='BIC')
plt.plot(n_components, [m.aic(dataframe) for m in models], label='AIC')
plt.legend(loc='best')
plt.xlabel('n_components');
```

```python
gmm = GaussianMixture(n_components=3)
gmm.fit(dataframe)
```
