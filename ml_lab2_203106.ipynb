```python
from sklearn.datasets import fetch_california_housing
california_housing = fetch_california_housing(as_frame=True)
```

```python
california_housing
```

```python
import pandas as pd
```

```python
california_housing.target
```

```python
california_housing.data
```

```python
dataframe = pd.DataFrame(data=california_housing.data, columns=california_housing.feature_names)
dataframe.head()
```

```python
dataframe["Median_house_value"] = california_housing.target
```

```python
dataframe.head()
```

```python
from seaborn import pairplot
pairplot(dataframe)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,10))
sns.heatmap(dataframe.corr(),cmap='coolwarm',annot=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

normalized_data = scaler.fit_transform(dataframe)

print(normalized_data)
```

```python
normalized_dataframe = pd.DataFrame(data=normalized_data, columns = dataframe.columns)
```

```python
normalized_dataframe
```

```python
plt.figure(figsize=(8,5))
plt.hist(normalized_dataframe['Median_house_value'],bins=50,edgecolor='k')
plt.ylabel('Count',fontsize=15)
plt.xlabel('Median house value',fontsize=15)
plt.title("Histogram of target column",fontsize=18)
plt.show()
```

```python
normalized_dataframe['Median_house_value'].value_counts()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = normalized_dataframe.drop(['Median_house_value'],axis=1)
```

```python
X
```

```python
Y = normalized_dataframe['Median_house_value']
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
X_train
```

```python
Y_train
```

```python
regressor.fit(X_train,Y_train)
```

```python
predictions = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse = mean_squared_error(Y_test,predictions)
```

```python
print("MSE for linear regression: ",mse)
```

```python
from sklearn.linear_model import LassoCV,RidgeCV
```

```python
lassoCV = LassoCV(n_alphas=20)
```

```python
lassoCV.fit(X_train,Y_train)
```

```python
predictions = lassoCV.predict(X_test)
```

```python
mse = mean_squared_error(Y_test,predictions)
```

```python
print("MSE for LassoCV: ",mse)
```

```python
ridgeCV = RidgeCV(alphas=25)
```

```python
ridgeCV.fit(X_train,Y_train)
```

```python
predictions = ridgeCV.predict(X_test)
```

```python
mse = mean_squared_error(Y_test,predictions)
```

```python
print("MSE for RidgeCV: ",mse)
```
